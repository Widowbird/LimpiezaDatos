```python
import pandas as pd
#Cargo los archivos para analizar que informaci√≥n contienene cada uno de ellos
df1 = pd.read_excel("pcma_20251217-oficio-4770_2013.xlsx")
df2 = pd.read_excel("pcmav-alto-estandar_20251217_oficio-4770_2013.xlsx")
df3 = pd.read_excel("pcmav-estandar-normal_20251217_oficio-4770_2013 (1).xlsx")
```

```python
#Revisar archivos
df1.head(20)
```

```python
#Revisar archivos
df2.head(20)
```

```python
#Revisar archivos
df3.head(20)
```

```python
#Limpio archivo 1
servicios = df1.iloc[3:5, 0]
servicios = ",".join(servicios.dropna().astype(str))
df1_clean = pd.read_excel("pcma_20251217-oficio-4770_2013.xlsx",header=9)
df1_clean['Descripcion'] = "Punto Bip"
df1_clean['servicios'] = servicios
df1_clean.head()
```

```python
#Limpio archivo 2
servicios2 = df2.iloc[3:9, 0]
servicios2 = ",".join(servicios2.dropna().astype(str))
df2_clean = pd.read_excel("pcmav-alto-estandar_20251217_oficio-4770_2013.xlsx",header=13)
df2_clean['Descripcion'] = "Alto"
df2_clean['servicios'] = servicios2
#print(servicios2)
df2_clean.head(20)
```

```python
#Limpio archivo 3
servicios3 = df3.iloc[3:7, 0]
servicios3 = ",".join(servicios3.dropna().astype(str))
df3_clean = pd.read_excel("pcmav-estandar-normal_20251217_oficio-4770_2013 (1).xlsx",header=10)
df3_clean['Descripcion'] = "Normal"
df3_clean['servicios'] = servicios3
df3_clean.head()
```

```python
#Me percato que en el archivo 1 existe una columna que no aparece los otros archivos, "Nombre de Fantasia", procedo a eliminar y anexar archivos
df1_clean = df1_clean.drop(columns=['NOMBRE DE FANTASIA '])
df_final = pd.concat(
    [df1_clean, df2_clean, df3_clean],
    ignore_index=True
)
df_final.head()
```

```python
#Agrego columnas Pais y Concateno la Latitud y Longitud
df_final['PAIS'] = 'Chile'
df_final['LatLon'] = (
    df_final['LATITUD'].astype(str)
    + ',' +
    df_final['LONGITUD'].astype(str)
)
df_final.head(20)
```

```python
#exporta mi archivo para usarlo
df_final.to_excel(
    "df_final_chile_bip.xlsx",
    index=False
)
```
